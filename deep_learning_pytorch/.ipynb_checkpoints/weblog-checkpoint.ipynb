```python
import pandas as pd
def extract_implicit_feedbacks(filename):
  #df = pd.DataFrame.from_csv(filename, sep='\t',header=None,index_col=None)
  df = pd.read_excel('merged_weblogs.xlsx')
  #df.columns = headers
  df2 = df[df.user_server == 'DAA']
  n,_ = df2.shape
  if n == 0:
    return ([],[])
  df3 = df2.dropna(axis='columns',how='all')
  keepcols = ['click_action','click_context','evar7','evar9','evar28','evar40','evar50','date_time']
  groupby_key = 'evar18'
 
  search_term=None
  search_scope=""
  search_timestamp=""
  
  search_feedbacks=[]
  no_clicks=[]
  previous_search_term=None
  for name,group in df3[keepcols].groupby(df3[groupby_key]):
    for index, row in group.iterrows():
      if row['evar50']==row['evar50']:
        user_id= row['evar50']

      if row['evar7']==row['evar7']:
        search_term=row['evar7']
        
        #print(search_term)
        search_timestamp = row['date_time']
        if previous_search_term is not None:
          previous_search_term = None
          no_clicks.append({
              'search_term':search_term,
              'date_time' : search_timestamp,
              'user_id' : user_id,
              'search_scope' : search_scope,          
              })
    
        
      if row['evar9']==row['evar9']:
        search_scope=row['evar9']
      click_action = row['click_action']
      if click_action == click_action and click_action != "javascript:void(0)":
        previous_search_term = search_term
        search_feedbacks.append({
            'search_term' : search_term,
            'search_scope' : search_scope,          
            'user_id' : user_id,
            'date_time' : search_timestamp,
            'click_context' : row['click_context'],
            'clicked': click_action          
            })
    
  return search_feedbacks,no_clicks, df2
  
  
```
